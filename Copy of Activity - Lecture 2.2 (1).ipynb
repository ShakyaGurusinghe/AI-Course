```python
fact = int(input("Enter a number :"))


def Fac(factorial) :
    result = 1
    while factorial >= 1 :
            result = result * factorial
            factorial = factorial - 1
    return result


print("Factorial of ",fact," is ",Fac(fact))
```

b) Create a python program to input marks of 10 students and print the maximum mark, minimum mark and average mark of the students.



```python
i = 0
summation = 0
myList = []

# input marks of 10 students
while i < 10:
    marks = int(input("Enter the marks for student{}: ".format(i + 1)))
    myList.append(marks)
    summation = summation + marks
    i = i + 1

# calculate max, min, and average marks of 10 students
maximum_mark = max(myList)
minimum_mark = min(myList)
avrg = summation / 10

# display max, min, and average marks of 10 students
print("Maximum mark = ", maximum_mark)
print("Minimum mark = ", minimum_mark)
print("Average mark = ", avrg)
```

c)Create a program to input five marks of a student and display the grades.

• Mark > 75 – A

• Mark 65 to 75 – B

• Mark 55 to 64 – C

• Mark 45 to 54 – S

• Mark < 45 – F

```python
i = 0

while i<5:
    marks = int(input("Enter the marks of the students : "))

    if marks>=75 :
        print("Grade=A")
    elif 65<= marks <75:
        print("Grade=B")
    elif 55 <= marks <65:
        print("Grade=C")
    elif 45 <= marks < 55:
        print("Grade=S")
    elif marks<45:
        print("Grade=F")
    else:
        print("You entered marks are invalid!")

    i = i + 1
```

d)  Write a program to input a series of numbers terminated by -999. Calculate and print the sum of the numbers entered.

```python
i = 0
sum = 0

while True:
    number = int(input("Enter a number:"))
    if number== -999:
        break
    sum = sum + number

    i = i + 1

print("Sum = ",sum)
```
